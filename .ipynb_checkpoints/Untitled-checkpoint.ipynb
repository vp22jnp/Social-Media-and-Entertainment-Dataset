# Whic platform used the most

```python
ax=sns.countplot(data=df , x='Primary Platform', color='gold')

ax.bar_label(ax.containers[0])
```

# What Devices used the most for entertainment

```python
ax=sns.countplot(data=df , x='Preferred Device for Entertainment', color='cyan')

ax.bar_label(ax.containers[0])
```

```python
df.head()
```

# Most platform used by country

```python
Most_platform_used_by_country= df.groupby('Country')['Primary Platform'].value_counts().sort_values()
plt.figure(figsize=(20,5))
ax=Most_platform_used_by_country.plot(kind='bar',color=['red','green','teal','olive','gold'])

ax.bar_label(ax.containers[0])
plt.title('Most_platform_used_by_country')
```

# Top 10 Platfrom users by Country

```python
Most_platform_used_by_country = df.groupby('Country')['Primary Platform'].value_counts()
Most_platform_used_by_country = Most_platform_used_by_country.sort_values(ascending=False)
Top_10_platforms = Most_platform_used_by_country.head(10)
plt.figure(figsize=(20, 5))
ax = Top_10_platforms.plot(kind='bar', color='teal') 
ax.bar_label(ax.containers[0])  
plt.title('Top 10 Most Used Platforms by Country')
plt.xlabel('Country - Platform')
plt.ylabel('Counts')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

# Time given on platform for entertainment purpose

```python
Daily_Entertainment_Time_by_country= df.groupby('Country')['Daily Entertainment Time (hrs)'].mean().sort_values()
plt.figure(figsize=(15,5))
ax=Daily_Entertainment_Time_by_country.plot(kind='bar', color=['red','green','teal','olive','gold'])

ax.bar_label(ax.containers[0])
plt.title('Daily_Entertainment_Time_by_country')
```

# Average Daily Messaging Time hrs by country

```python
Daily_Messaging_Time_hrs_by_country= df.groupby('Country')['Daily Messaging Time (hrs)'].mean().sort_values()
plt.figure(figsize=(15,5))
ax=Daily_Messaging_Time_hrs_by_country.plot(kind='bar',color=['red','green','teal','olive','gold'])

ax.bar_label(ax.containers[0])
plt.title('Daily_Messaging_Time_hrs_by_country')
```

```python
df.head()
```

# Average Time on Educational Platforms by country

```python
Time_on_Educational_Platforms_by_country= df.groupby('Country')['Time on Educational Platforms (hrs)'].mean().sort_values()
plt.figure(figsize=(15,5))
ax=Time_on_Educational_Platforms_by_country.plot(kind='bar',color=['red','green','teal','olive','gold'])

ax.bar_label(ax.containers[0])
plt.title('Time on Educational Platforms (hrs)')
```

# Average users age countrywise to use platforms

```python
heatmap_data = df.groupby(['Country', 'Primary Platform'])['Age'].mean().unstack()
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'Age'})
```

# Top 10 Ages used for educational purpose

```python
Top_10_age_used_on_educational_platforms = df.groupby('Age')['Time on Educational Platforms (hrs)'].mean()
Top_10_age_used_on_educational_platforms = Top_10_age_used_on_educational_platforms.sort_values(ascending=False)
Top_10_age_edu = Top_10_age_used_on_educational_platforms.head(10)
plt.figure(figsize=(20, 5))
ax = Top_10_age_edu.plot(kind='bar', color='teal') 
ax.bar_label(ax.containers[0])  
plt.title('Top_10_age_used_on_educational platforms')
plt.xlabel('Age')
plt.ylabel('Avg Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

```python
Top_10_age_used_social_media_for_entertainement = df.groupby('Age')['Daily Social Media Time (hrs)'].mean()
Top_10_age_used_social_media_for_entertainement = Top_10_age_used_social_media_for_entertainement.sort_values(ascending=False)
Top_10_age = Top_10_age_used_social_media_for_entertainement.head(10)
plt.figure(figsize=(20, 5))
ax = Top_10_age.plot(kind='bar', color='teal') 
ax.bar_label(ax.containers[0])  
plt.title('Top_10_age_used_social_media_for_entertainement')
plt.xlabel('Age')
plt.ylabel('Avg Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```
